Model does not understand the categorical data model only understand mathematical equationa and numerical values. So the process of converting categorical data into numerical value is know as `Data Encoding`.

### Data Encoding Techniques 

1. Nominal/OHE Encoding
2. Label and Ordinal Encoding
3. Target Guided Ordinal Encoding

### Nominal/OHE Encoding :

One hot encoding, also known as nominal encoding is a technique used to represent categorical data as numerical data, which is more suitable for machine learning algorithms.
In this technique, each category is represented as a binary vector where each bit corresponds to a unique category. For example, if we have a categorical variable "color" with three possible values (red,green,blue), we can represent it as one hot encoding as follows:

1. Red: [1,0,0]
2. Green : [0,1,0]
3. Blue : [0,0,1]

### Disadvantages of OHE :

1. If one of categories have 100 unique different categories than OHE create 100 few features which increase the number of cloumns and for that we required more data.
2. Sparse Matrix - number of 1 and 0 are huge in features of datasets it leads to overfitting problem.

```python
import pandas as pd
from sklearn.preprocessing import OneHotEncoder 
```

```python
## Create a simple dataframe

df= pd.DataFrame({
    'color' : ['red','blue','green','red','blue']

})
```

```python
df.head()
```

```python
## Create an instance of OneHotencoder
encoder = OneHotEncoder()
```

```python
## perform fit and transform
encoded = encoder.fit_transform(df[['color']]).toarray()
```

* Here doing alphabatically sorting

```python
import pandas as pd
encoder_df = pd.DataFrame(encoded, columns = encoder.get_feature_names_out())
```

```python
encoder_df
```

```python
## for new data
encoder.transform([['blue']]).toarray()
```

```python
pd.concat([df,encoder_df],axis=1)
```

### Label Encoding

* Label encoding and ordinal encoding are two technique used to encode categorical data as numerical data.

* Label encoding involves assigning a unique numerical label to each category in the variable. The labels are usually assigned in alphabetical order or based on the frequency of the categories. For example, if we have a categorical variable "color" with three possible values (red,green,blue), we can represent it using label encoding as follows:

1. Red : 1
2. Green : 2
3. Blue : 3

### Disadvantages of LabelEncoder :

* Model start thinking that numerical value which are higher will greater than other, it start comaprison with others values.
* It start assigning rank, it suitable case where we required ranking in our data.

```python
from sklearn.preprocessing import LabelEncoder
lbl_encoder = LabelEncoder()
```

```python
lbl_encoder.fit_transform(df[['color']])
```

```python
lbl_encoder.transform([['red']])
```

```python
lbl_encoder.transform([['blue']])
```

```python
lbl_encoder.transform([['green']])
```

### Ordinal Encoding :

It is used to encode categorical data that have an intrinsic order or ranking. In this technique, each category is assigned a numerical value based on its position in the order. For example, if we have a categorical variable "education level" with four possible values (high school, college, graduate, post-graduate), we can represent it using ordinal encoding as follows:

1. High School : 1
2. College : 2
3. Graduate : 3
4. Post- Graduate : 4

```python
from sklearn.preprocessing import OrdinalEncoder
```

```python
## Create a sample dataframe with an ordinal variable

df = pd.DataFrame({
    'size' : ['small','medium','large','medium','small','large']

})
```

```python
df
```

```python
## Create an insatnce of ordinalencoder and then fit_transform
encoder = OrdinalEncoder(categories = [['small','medium','large']])
```

```python
encoder.fit_transform(df[['size']])
```

```python
encoder.transform([['small']])
```

### Target Guided Ordinal Encoding :

It is a technique used to encode categorical variables based on their relationship with the target variables. This encoding technique is useful when we have a categorical variable with a large number of unique categories, and we want to use this variable as a freature in our machine learning model.

In Target Guided Ordinal Encoding, we replace each categorical variable with a numerical value based on the mean or median of the Target variable for that category. This creates a monotonic relationship between the categorical variable and the target variable, which can improve the predictive power of our model.

```python
import pandas as pd

# create a sample dataframe with a categorical variable and a target variable

df= pd.DataFrame({
    'city' : ['New York','London','Paris','Tokyo','New York','Paris'],
    'price' : [200,150,300,250,180,320]

})
```

```python
df
```

```python
df.groupby('city')['price'].mean()
```

```python
mean_price = df.groupby('city')['price'].mean().to_dict()
```
